# Basics of reguler expression

`^`: to start with  
`$`: to end with  
`\d`:single numerical digit  
`\s`:space  
`\w`:any digit any letter Lower or upper case  
`[abcd]quantifier`: this is possible sets of letters with given quantifier  
`[abcd]{2}` possible outcome `ac`,`bd .....  

sets: you can create a custom set `[a-zA-Z!]` all lower upper case letter and exclamation mark  

## Quantifier

|Quantifier|Meaning|Example|Matches...|
|---|---|---|---|
|`*`|0 or more times|`a*`|`""`, `"a"`, `"aaa"`|
|`+`|1 or more times|`a+`|`"a"`, `"aa"`, but not `""`|
|`?`|0 or 1 time (optional)|`a?`|`""`, `"a"`|
|`{n}`|Exactly **n** times|`a{3}`|`"aaa"`|
|`{n,}`|**n or more** times|`a{2,}`|`"aa"`, `"aaa"`, `"aaaa"`|
|`{n,m}`|Between **n and m** times|`a{2,4}`|`"aa"`, `"aaa"`, `"aaaa"`|

## Groups

you create a group by adding parentheses to the regular expression `(\w*)\s(\w*)` this will give use two groups maybe one for last name and the other for the first name.

### how to name a group


You can name a group using this:`?P<name_of he group>`

<span style="color:rgb(255, 0, 0)">Example</span> :

### Processing Function

````python
def process_image(image_path, save_path):
    img = Image.open(image_path)

    # Brightening
    brightener = ImageEnhance.Brightness(img)
    img = brightener.enhance(1.5)

    # Grayscale
    gray = img.convert("L")

    # Binarization
    def threshold_pixel(p):
        return 255 if p > 128 else 0

    binary = gray.point(threshold_pixel)

    # Save result
    binary.save(save_path)
````

## Now Transfroming all the Pictrues into Text

````python
import os
from PIL import Image, ImageFilter
import pytesseract

# Configuration
input_folder = "Processed_images"        # Folder with preprocessed images
output_text_folder = "text_ocr"          # Folder to save text files

# Define your zones (left, top, right, bottom)
ZONES = {
    "header": (0, 20, 1600, 280),
    "seller_info": (0, 281, 674, 750),
    "client_info": (674, 281, 1600, 750),
    "body_content": (0, 750, 1600, 2290)
}

def preprocess_image(image):
    """Enhances image for better OCR results"""
    image = image.convert('L')  # Convert to grayscale
    image = image.filter(ImageFilter.SHARPEN)  # Sharpen text
    image = image.point(lambda x: 0 if x < 140 else 255)  # Binarize
    return image

def extract_zone_text(img, zone_coords):
    """Extracts and cleans text from a specific zone"""
    zone_img = img.crop(zone_coords)
    zone_img = preprocess_image(zone_img)
    
    # Use appropriate OCR config based on zone content
    if zone_coords == ZONES["header"]:
        config = '--psm 6 -c tessedit_char_whitelist=0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ/-'
    else:
        config = '--psm 6'  # Assume paragraph text for other zones
    
    text = pytesseract.image_to_string(zone_img, config=config)
    return text.strip()

def process_image(image_path):
    """Processes a single image through all zones"""
    with Image.open(image_path) as img:
        extracted_text = ""
        
        for zone_name, coords in ZONES.items():
            zone_text = extract_zone_text(img, coords)
            
            # Format the extracted text
            extracted_text += f"\n\n=== {zone_name.upper()} ===\n{zone_text}"
        
        return extracted_text.strip()

# Main processing loop
for filename in os.listdir(input_folder):
    if filename.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".tiff")):
        image_path = os.path.join(input_folder, filename)
        
        # Create output folder if it doesn't exist
        os.makedirs(output_text_folder, exist_ok=True)
        
        # Generate output text path
        text_filename = os.path.splitext(filename)[0] + ".txt"
        text_path = os.path.join(output_text_folder, text_filename)
        
        # Process and save text
        organized_text = process_image(image_path)
        with open(text_path, "w", encoding="utf-8") as f:
            f.write(organized_text)

print(f"✅ OCR complete. Organized text files saved in: {output_text_folder}")
````

# CSV File

````python
import re
import os
import pandas as pd

# Define all regex patterns (removed vat and gross_worth patterns)
regex_patterns = {
    'invoice_number': r'(\d{8})\n',
    'date': r'\d{1,2}/\d{2}/\d{4}',
    'net_worth': r'Total\s*\$\s*([\d\s]+,\d{2})',  # Will take first match
    'tax_id_1': r'Tax Id:\s*(\d+-\d{2,3}-\d+)',  # First match
    'tax_id_2': r'Tax Id:\s*(\d+-\d{2,3}-\d+)',  # Second match
    'iban': r'IBAN:\s*(\w+)',
    'seller_city_state_zip': r'\n([a-zA-Z\s]+),?\s*([A-Z]{2})\s*(\d{5})\n',  # For seller
    'client_city_state_zip': r'\n([a-zA-Z\s]+),?\s*([A-Z]{2})\s*(\d{5})\n',  # For client
    'seller_name_address': r'Seller:\n+([A-Za-z\s,-]+)\n+([A-Za-z0-9\s,.]+)\n[a-zA-Z\s]+,?\s*[A-Z]{2}\s*\d{5}',
    'client_name_address': r'Client:\n+([A-Za-z\s,-]+)\n+([A-Za-z0-9\s,.]+)\n[a-zA-Z\s]+,?\s*[A-Z]{2}\s*\d{5}'
}

def extract_data(raw_text):
    """Extract all data from raw text using the regex patterns"""
    data = {}
    
    # Invoice number
    match = re.search(regex_patterns['invoice_number'], raw_text)
    data['invoice_number'] = match.group(1) if match else None
    
    # Date
    match = re.search(regex_patterns['date'], raw_text)
    data['date'] = match.group(0) if match else None
    
    # Net worth only (vat and gross_worth will be calculated later)
    financial_matches = re.findall(regex_patterns['net_worth'], raw_text)
    if len(financial_matches) > 0:
        # First remove any whitespace, then replace comma with dot
        cleaned_value = financial_matches[0].strip().replace(" ", "").replace(",", ".")
        data['net_worth'] = float(cleaned_value)
    
    # Tax IDs
    tax_matches = re.findall(regex_patterns['tax_id_1'], raw_text)
    data['tax_id_1'] = tax_matches[0] if len(tax_matches) > 0 else None
    data['tax_id_2'] = tax_matches[1] if len(tax_matches) > 1 else None
    
    # IBAN
    match = re.search(regex_patterns['iban'], raw_text)
    data['iban'] = match.group(1) if match else None
    
    # Location info - seller
    loc_matches = re.findall(regex_patterns['seller_city_state_zip'], raw_text)
    if loc_matches:
        data['seller_city'] = loc_matches[0][0] if len(loc_matches) > 0 else None
        data['seller_state'] = loc_matches[0][1] if len(loc_matches) > 0 else None
        data['seller_zip'] = loc_matches[0][2] if len(loc_matches) > 0 else None
    
    # Location info - client
    loc_matches = re.findall(regex_patterns['client_city_state_zip'], raw_text)
    if loc_matches:
        data['client_city'] = loc_matches[0][0] if len(loc_matches) > 0 else None
        data['client_state'] = loc_matches[0][1] if len(loc_matches) > 0 else None
        data['client_zip'] = loc_matches[0][2] if len(loc_matches) > 0 else None
    
    # Seller name and address
    # First try the general regex
    match = re.findall(regex_patterns['seller_name_address'], raw_text)
    if not match:
        # If no match, try the niche regex
        match = re.findall(r'Seller:\n+([A-Za-z\s,-]+)\n+([A-Za-z0-9\s,.]+)\n[a-zA-Z\s]+,' , raw_text)

    data['seller_name'] = match[0][0] if match else None
    data['seller_address'] = match[0][1] if match else None

    # Same approach for client
    match = re.findall(regex_patterns['client_name_address'], raw_text)
    if not match:
        match = re.findall(r'Client\n+([A-Za-z\s,-]+)\n+([A-Za-z0-9\s,.]+)\n[a-zA-Z\s]+,', raw_text)

    data['client_address'] = match[0][1] if match else None
    data['client_name'] = match[0][0] if match else None
    
    return data

def process_folder(folder_path):
    """Process all text files in folder and return DataFrame"""
    all_data = []
    
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            filepath = os.path.join(folder_path, filename)
            with open(filepath, 'r', encoding='utf-8') as file:
                raw_text = file.read()
                extracted = extract_data(raw_text)
                extracted['source_file'] = filename
                all_data.append(extracted)
    
    # Create DataFrame
    df = pd.DataFrame(all_data)
    
    # Calculate VAT (10% of net worth) and gross worth (net + VAT)
    df['vat'] = df['net_worth'] * 0.1
    df['gross_worth'] = df['net_worth'] + df['vat']
    
    return df

# Main execution
if __name__ == "__main__":
    folder_path = 'text_ocr'  # Change this if needed
    df = process_folder(folder_path)
    df = df.replace('\n', '', regex=True)# to remove all \n

    
    # Check for missing values
    print("\nMissing Values Summary:")
    print(df.isna().sum())
    
    # Optionally save to CSV
    df.to_csv('extracted_invoice_data.csv', index=False)
    print("\nData saved to 'extracted_invoice_data.csv'")
````

# jason file

````python
import re
import os
import json

# Define all regex patterns
regex_patterns = {
    'invoice_number': r'(\d{8})\n',
    'date': r'\d{1,2}/\d{2}/\d{4}',
    'tax_id': r'Tax Id:\s*(\d+-\d{2,3}-\d+)',
    'iban': r'IBAN:\s*(\w+)',
    'city_state_zip': r'\n([a-zA-Z\s]+),?\s*([A-Z]{2})\s*(\d{5})\n',
    'name_address': r'(?:Seller|Client):\n+([A-Za-z\s,-]+)\n+([A-Za-z0-9\s,.]+)\n[a-zA-Z\s]+,?\s*[A-Z]{2}\s*\d{5}',
    'items': r'(\d+)\.\s+([^\n]+)\s+([\d,]+)\s+each\s+([\d,]+)\s+([\d,]+)\s+(\d+%)',
    'summary': r'SUMMARY\nVAT\s*\[%\]\s*Net worth\s*VAT\s*Gross worth\n(\d+%)\s*([\d,]+)\s*([\d,]+)\s*([\d,]+)',
    'net_worth': r'Total\s*\$\s*([\d\s]+,\d{2})'
}

def clean_text(text):
    """Remove newlines and extra spaces from text"""
    if isinstance(text, str):
        return ' '.join(text.replace('\n', ' ').split())
    return text

def extract_items(raw_text):
    """Extract items from the invoice body"""
    items = []
    matches = re.finditer(regex_patterns['items'], raw_text)
    
    for match in matches:
        item = {
            'item_no': match.group(1),
            'description': clean_text(match.group(2)),
            'quantity': float(match.group(3).replace(',', '.')),
            'unit_price': float(match.group(4).replace(',', '.')),
            'net_price': float(match.group(5).replace(',', '.')),
            'vat_percentage': match.group(6)
        }
        items.append(item)
    
    return items

def extract_summary(raw_text):
    """Extract summary information"""
    match = re.search(regex_patterns['summary'], raw_text)
    if match:
        return {
            'vat_percentage': match.group(1),
            'net_worth': float(match.group(2).replace(',', '.')),
            'vat_amount': float(match.group(3).replace(',', '.')),
            'gross_worth': float(match.group(4).replace(',', '.'))
        }
    return None

def extract_data(raw_text):
    """Extract all data from raw text using the regex patterns"""
    data = {
        'header': {},
        'seller_info': {},
        'client_info': {},
        'items': [],
        'summary': {}
    }
    
    # Header information
    match = re.search(regex_patterns['invoice_number'], raw_text)
    data['header']['invoice_number'] = clean_text(match.group(1)) if match else None
    
    match = re.search(regex_patterns['date'], raw_text)
    data['header']['date'] = match.group(0) if match else None
    
    # Seller information
    seller_matches = re.findall(regex_patterns['name_address'], raw_text)
    if seller_matches:
        data['seller_info']['name'] = clean_text(seller_matches[0][0])
        data['seller_info']['address'] = clean_text(seller_matches[0][1])
    
    loc_matches = re.findall(regex_patterns['city_state_zip'], raw_text)
    if loc_matches and len(loc_matches) > 0:
        data['seller_info']['city'] = clean_text(loc_matches[0][0])
        data['seller_info']['state'] = loc_matches[0][1]
        data['seller_info']['zip'] = loc_matches[0][2]
    
    tax_matches = re.findall(regex_patterns['tax_id'], raw_text)
    if tax_matches and len(tax_matches) > 0:
        data['seller_info']['tax_id'] = tax_matches[0]
    
    match = re.search(regex_patterns['iban'], raw_text)
    data['seller_info']['iban'] = match.group(1) if match else None
    
    # Client information
    if seller_matches and len(seller_matches) > 1:
        data['client_info']['name'] = clean_text(seller_matches[1][0])
        data['client_info']['address'] = clean_text(seller_matches[1][1])
    elif len(seller_matches) == 1:
        client_matches = re.findall(r'Client:\n+([A-Za-z\s,-]+)\n+([A-Za-z0-9\s,.]+)', raw_text)
        if client_matches:
            data['client_info']['name'] = clean_text(client_matches[0][0])
            data['client_info']['address'] = clean_text(client_matches[0][1])
    
    if loc_matches and len(loc_matches) > 1:
        data['client_info']['city'] = clean_text(loc_matches[1][0])
        data['client_info']['state'] = loc_matches[1][1]
        data['client_info']['zip'] = loc_matches[1][2]
    
    if tax_matches and len(tax_matches) > 1:
        data['client_info']['tax_id'] = tax_matches[1]
    
    # Items and summary
    data['items'] = extract_items(raw_text)
    data['summary'] = extract_summary(raw_text)
    
    # Fallback for net worth if summary not found
    if not data['summary']:
        match = re.search(regex_patterns['net_worth'], raw_text)
        if match:
            net_worth = float(match.group(1).replace(' ', '').replace(',', '.'))
            data['summary'] = {
                'vat_percentage': '10%',
                'net_worth': net_worth,
                'vat_amount': net_worth * 0.1,
                'gross_worth': net_worth * 1.1
            }
    
    return data

def process_folder(folder_path):
    """Process all text files in folder and return list of JSON data"""
    all_data = []
    
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            filepath = os.path.join(folder_path, filename)
            with open(filepath, 'r', encoding='utf-8') as file:
                raw_text = file.read()
                extracted = extract_data(raw_text)
                extracted['source_file'] = filename
                all_data.append(extracted)
    
    return all_data

class NoNewlineEncoder(json.JSONEncoder):
    """Custom JSON encoder to ensure no newlines in output"""
    def encode(self, obj):
        result = super().encode(obj)
        return result.replace('\\n', ' ')

if __name__ == "__main__":
    folder_path = 'text_ocr'  # Change this to your folder path
    invoice_data = process_folder(folder_path)
    
    # Save to JSON file with clean formatting
    with open('invoices.json', 'w', encoding='utf-8') as f:
        json.dump(invoice_data, f, indent=2, ensure_ascii=False, cls=NoNewlineEncoder)
    
    print("\nData successfully saved to 'invoices.json'")
    print(f"Processed {len(invoice_data)} invoices")
````

# SQL database

````python
import sqlite3
from sqlite3 import Error
import pandas as pd

class InvoiceDatabase:
    def __init__(self, db_name='invoice_system.db'):
        self.db_name = db_name
        self.conn = None
        
    def create_connection(self):
        """Create a database connection to a SQLite database"""
        try:
            self.conn = sqlite3.connect(self.db_name)
            return self.conn
        except Error as e:
            print(f"Error connecting to database: {e}")
        return None
    
    def create_tables(self):
        """Create tables exactly matching the schema diagram"""
        try:
            c = self.conn.cursor()
            
            # Create Sellers table (left entity in diagram)
            c.execute('''CREATE TABLE IF NOT EXISTS Sellers (
                         tax_id TEXT PRIMARY KEY,
                         name TEXT NOT NULL,
                         address TEXT,
                         city TEXT,
                         state TEXT,
                         zip_code TEXT,
                         iban TEXT)''')
            
            # Create Clients table (right entity in diagram)
            c.execute('''CREATE TABLE IF NOT EXISTS Clients (
                         tax_id TEXT PRIMARY KEY,
                         name TEXT NOT NULL,
                         address TEXT,
                         city TEXT,
                         state TEXT,
                         zip_code TEXT)''')
            
            # Create Invoices table (center entity in diagram)
            c.execute('''CREATE TABLE IF NOT EXISTS Invoices (
                         invoice_number TEXT PRIMARY KEY,
                         date TEXT,
                         net_worth REAL,
                         vat REAL,
                         gross_worth REAL,
                         client_tax_id TEXT,
                         seller_tax_id TEXT,
                         FOREIGN KEY (client_tax_id) REFERENCES Clients (tax_id),
                         FOREIGN KEY (seller_tax_id) REFERENCES Sellers (tax_id))''')
            
            self.conn.commit()
        except Error as e:
            print(f"Error creating tables: {e}")
    
    def insert_from_csv(self, csv_file='extracted_invoice_data.csv'):
        """Insert data from CSV file into the database"""
        try:
            # Read CSV file
            df = pd.read_csv(csv_file)
            
            # Insert sellers data
            sellers_df = df[['tax_id_1', 'seller_name', 'seller_address', 
                           'seller_city', 'seller_state', 'seller_zip', 'iban']]
            sellers_df = sellers_df.drop_duplicates(subset=['tax_id_1'])
            for _, row in sellers_df.iterrows():
                self.conn.execute('''INSERT OR IGNORE INTO Sellers
                                  (tax_id, name, address, city, state, zip_code, iban)
                                  VALUES (?, ?, ?, ?, ?, ?, ?)''',
                                  tuple(row))
            
            # Insert clients data
            clients_df = df[['tax_id_2', 'client_name', 'client_address', 
                           'client_city', 'client_state', 'client_zip']]
            clients_df = clients_df.drop_duplicates(subset=['tax_id_2'])
            for _, row in clients_df.iterrows():
                self.conn.execute('''INSERT OR IGNORE INTO Clients
                                  (tax_id, name, address, city, state, zip_code)
                                  VALUES (?, ?, ?, ?, ?, ?)''',
                                  tuple(row))
            
            # Insert invoices data
            invoices_df = df[['invoice_number', 'date', 'net_worth', 'vat',
                             'gross_worth', 'tax_id_2', 'tax_id_1']]
            for _, row in invoices_df.iterrows():
                self.conn.execute('''INSERT OR REPLACE INTO Invoices
                                  (invoice_number, date, net_worth, vat, gross_worth,
                                   client_tax_id, seller_tax_id)
                                  VALUES (?, ?, ?, ?, ?, ?, ?)''',
                                  tuple(row))
            
            self.conn.commit()
            print(f"Successfully imported data from {csv_file}")
        except Exception as e:
            print(f"Error importing from CSV: {e}")
    
    def export_to_sql(self, filename='invoice_schema_dump.sql'):
        """Export database to SQL file matching the schema diagram"""
        try:
            with open(filename, 'w') as f:
                for line in self.conn.iterdump():
                    f.write(f'{line}\n')
            print(f"Database exported to {filename}")
        except Error as e:
            print(f"Error exporting database: {e}")
    
    def generate_dataframe(self, table_name):
        """Generate pandas DataFrame from a table"""
        try:
            return pd.read_sql_query(f"SELECT * FROM {table_name}", self.conn)
        except Error as e:
            print(f"Error generating DataFrame: {e}")
            return None
    
    def close_connection(self):
        """Close the database connection"""
        if self.conn:
            self.conn.close()

# Example usage
if __name__ == "__main__":
    # Initialize database
    db = InvoiceDatabase()
    db.create_connection()
    db.create_tables()
    
    # Import data from CSV
    db.insert_from_csv('extracted_invoice_data.csv')
    
    # Generate DataFrames to verify data
    sellers_df = db.generate_dataframe('Sellers')
    clients_df = db.generate_dataframe('Clients')
    invoices_df = db.generate_dataframe('Invoices')
    
    print("\nSellers Table:")
    print(sellers_df.head())
    print("\nClients Table:")
    print(clients_df.head())
    print("\nInvoices Table:")
    print(invoices_df.head())
    
    # Export database to SQL
    db.export_to_sql()
    
    # Close connection
    db.close_connection()
````
